```python
df2 = df[['Name', 'Score', 'Genres', 'Episodes', 'Premiered']]
df2.head(5)
```

```python
df2 = df2.assign(Premiered = lambda x: x['Premiered'].str.extract('(\d+)'))
df2.head(10)
```

```python
def to_float(x):
    try:
        x = float(x)
    except:
        x = np.nan
    return x
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return 0
```

```python
df2['Score'] = df2['Score'].apply(to_float).astype(float)
df2['Episodes'] = df2['Episodes'].apply(convert_int).astype(int)
df2['Premiered'] = df2['Premiered'].apply(convert_int).astype(int)
df2.info()
```

**The knowledge-based recommender**

```python
df2['Genres'] = df2['Genres'].str.strip('()').str.split(',')
df2
```

```python
import warnings
warnings.filterwarnings('ignore')

s = df2.apply(lambda x: pd.Series(x['Genres']), axis = 1).stack()\
.reset_index(level = 1, drop = True)
s.name = 'Genres'
gen_df = df2.drop('Genres', axis = 1).join(s)
gen_df['Genres'] = gen_df['Genres'].str.lower() 
gen_df.head(10)
```

```python
import copy

def build_chart(gen_df, percentile = 0.8):
    
    print('Input prefereed genres')
    Genres = input()
    print('Input shortest Episodes')
    low_Episodes = int(input())
    print('Input longest Episodes')
    high_Episodes = int(input())
    print('Input earliest year')
    low_year = int(input())
    print('Input latest year')
    high_year = int(input())
    movies = copy.deepcopy(gen_df)
    movies = movies[(movies['Genres'] == Genres) &
                    (movies['Episodes'] >= low_Episodes) &
                    (movies['Episodes'] <= high_Episodes) &
                    (movies['Premiered'] >= low_year) &
                    (movies['Premiered'] <= high_year)]
    q_movies = movies.sort_values('Score', ascending = False)
    return q_movies
```
